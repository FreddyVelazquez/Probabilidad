```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv("autos.csv")

# Asegurar que "price" sea numerico
df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

```python
# Estadisticos descriptivos
precio_min = df["price"].min()
precio_max = df["price"].max()
precio_moda = df["price"].mode()[0]
precio_media = df["price"].mean()
precio_mediana = df["price"].median()
precio_std = df["price"].std()
descripcion = df["price"].describe()

print("==== Estadisticos de Precios ====")
print(f"Valor minimo: {precio_min}")
print(f"Valor maximo: {precio_max}")
print(f"Moda: {precio_moda}")
print(f"Media: {precio_media}")
print(f"Mediana: {precio_mediana}")
print(f"Desviacion estandar: {precio_std}")
print("\nResumen con describe():\n", descripcion)

# Interpretacion
print("\n=== Interpretacion ===")
print("El precio minimo corresponde a los vehiculos mas economicos, mientras que el precio maximo muestra la existencia de automoviles de gama alta.")
print("La media y la mediana permiten ver la tendencia central; si la media es mayor que la mediana, indica que la distribucion esta sesgada a la derecha (precios muy altos influyen).")
print("La moda representa el precio que mas se repite, generalmente asociado a vehiculos de gama popular.")
print("La desviacion estandar nos indica la dispersion de los precios: un valor alto sugiere gran variabilidad en los costos de los automoviles.")
```

```python
# Histograma con regla de Sturges
n = df["price"].count()
bins_sturges = int(np.ceil(1 + np.log2(n)))

plt.figure(figsize=(8,5))
plt.hist(df["price"], bins=bins_sturges, edgecolor="black")
plt.title("Histograma de Precios de Automoviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

print(f"\nNumero optimo de bins (Sturges): {bins_sturges}")

# Interpretacion
print("\n=== Interpretacion Histograma ===")
print("El histograma muestra la distribucion de precios de los automoviles.")
print("Se observa que la mayoria de los vehiculos tienen precios en rangos bajos o medios, mientras que pocos modelos alcanzan precios muy altos.")
print("Esto confirma una distribucion asimetrica positiva (sesgo a la derecha).")
```

```python
# Boxplot Precio vs Body Style
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="body-style", y="price")
plt.title("Relacion entre Precio y Estilo de Carroceria (Body Style)")
plt.xlabel("Estilo de Carroceria")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()

print("\n=== Interpretacion Precio vs Body Style ===")
print("El diagrama de caja indica que los vehiculos tipo sedan y hatchback tienden a tener precios mas bajos en comparacion con los convertibles o hardtops.")
print("Los convertibles y carros de lujo muestran mayor dispersion y valores atipicos hacia precios altos.")
print("Esto refleja como el estilo de carroceria influye directamente en el valor del vehiculo.")
```

```python
# Boxplot Precio vs Drive Wheels
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="drive-wheels", y="price")
plt.title("Relacion entre Precio y Tipo de Traccion (Drive Wheels)")
plt.xlabel("Tipo de Traccion")
plt.ylabel("Precio")
plt.show()

print("\n=== Interpretacion Precio vs Drive Wheels ===")
print("Los autos con traccion delantera (fwd) suelen ser mas economicos y mas frecuentes.")
print("Los de traccion trasera (rwd) muestran mayor dispersion en precios, incluyendo modelos de lujo o deportivos.")
print("Los de traccion en las cuatro ruedas (4wd) aparecen con precios intermedios o altos, lo que refleja su mayor costo por tecnologia y desempeno.")
```

```python
# Boxplot Precio vs Engine Location
plt.figure(figsize=(6,5))
sns.boxplot(data=df, x="engine-location", y="price")
plt.title("Relacion entre Precio y Ubicacion del Motor (Engine Location)")
plt.xlabel("Ubicacion del Motor")
plt.ylabel("Precio")
plt.show()

print("\n=== Interpretacion Precio vs Engine Location ===")
print("Los autos con motor en la parte frontal (front) son los mas comunes y suelen tener precios moderados.")
print("En cambio, los pocos autos con motor en la parte trasera (rear) muestran precios considerablemente mas altos.")
print("Esto se asocia a vehiculos deportivos o de diseno especializado, que incrementan el costo.")
```

```python
# Conclusiones Generales
print("\n=== Conclusiones Generales de la Practica ===")
print("1. La mayoria de los automoviles se concentran en precios bajos o medios, con pocos modelos de precios muy altos.")
print("2. El mercado presenta una distribucion sesgada positivamente, influenciada por los autos de lujo.")
print("3. El estilo de carroceria, la traccion y la ubicacion del motor son factores que afectan significativamente el precio.")
print("4. Los autos sedan y hatchback suelen ser mas accesibles, mientras que los convertibles, deportivos y de traccion trasera son mas costosos.")
print("5. La practica permitio identificar como las variables categoricas influyen en el valor economico de los automoviles.")
```
