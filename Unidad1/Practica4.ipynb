### **Practica 4. Diagrama de Cajas y bigotes.**

```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1= []
for i in range (100):
  numero_aleatorio = round( random.uniform(50,100), 2 )
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2= []
for i in range (100):
  numero_aleatorio = round( random.uniform(40,93), 2 )
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
plt.boxplot(fabrica1)
plt.ylabel("Produccion en Metros.")
plt.title("Produccion de Fabrica 1")
plt.show()
```

```python
plt.boxplot(fabrica2)
plt.ylabel("Produccion en Metros.")
plt.title("Produccion de Fabrica 2")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1,
                     vert=True,
                     patch_artist=True)
ax1.set_title('Produccion de Fabrica 1')
ax1.set_ylabel('Produccion en Metros')
bplot2= ax2.boxplot(fabrica2,
                     vert=True,
                     patch_artist=True)
ax2.set_title('Produccion de Fabrica 2')
ax2.set_ylabel('Produccion en Metros')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1,
                     vert=True,
                     patch_artist=True)
ax1.set_title('Produccion de Fabrica 1')
ax1.set_ylabel('Produccion en Metros')
ax1.set_ylim(0,160)
bplot2= ax2.boxplot(fabrica2,
                     vert=True,
                     patch_artist=True)
ax2.set_title('Produccion de Fabrica 2')
ax2.set_ylabel('Produccion en Metros')
ax2.set_ylim(0,160)
plt.show()
```
