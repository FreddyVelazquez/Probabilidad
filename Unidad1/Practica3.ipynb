### **Practica 3. Tabla de frecuencias y regla de Sturges**

```python
#Basado en el libro de estadistica para la administracion pagina12
#Creamos una lisra vacia que representa la edad de 100 personas
import random
edades = []
#Hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range(100):
  #generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  #numero_aleatorio
  numero_aleatorio = random.randint(18,70)
  #agregamos el numero en la lista de valores
  edades.append(numero_aleatorio)
#imprimimos la lista de edades
print(edades)
```

```python
import pandas as pd
```

```python
#Crear un dataframe de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=['Edades'])
df.head()
```

```python
#Calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edades'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edades', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edades')

#imprimir la tabla de distribucion de frecuencias
print("Tabla de distribucion de frecuencias:")
tabla_frecuencias
```

CALCULAR LA FRECUENCIA RELATIVA

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

CALCULAR LA FRECUENCIA PORCENTUAL


```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frec Relativa'] * 100
tabla_frecuencias
```

CALCULAR LA FRECUENCIA ACUMULADA


```python
tabla_frecuencias['Frecuencia Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

#REGLA DE STURGES
1.Es una regla que sirve para calcular el numero de clases o intervalos idoneos en los que se debe de dividir un conjunto de datos
2.Establece que el nnumero de clases es igual a uno mas el logaritmo en base de dos del  numero total de datos
#C=1+*LOG* (N)
Donde N es el tamao de la muestra

```python
import math
```

```python
print(edades)
```

```python
#crear un DataFrame con las edades
df = pd.DataFrame({"Edad ": edades})
df
```

```python
#Calcular el numero de intervalos utilizando la regla de sturges
n = len(edades)
K = int(1+ math.log2(n))
print("Valor de n =",n,"\nNumero de Intervalos=",K)
```

```python
#Crear la tabla de Frecuencia
tabla_frecuencias = pd.cut(df['Edad '], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns =["Clase","Frecuencia"]
tabla_frecuencias
```

```python
#Ordenar la tabla de frecuencia por el intervalo
```

```python
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
#Generar columnas siguientes
tabla_frecuencias["Frec Rel"] = tabla_frecuencias["Frecuencia"] / n
tabla_frecuencias["Frec Porcent"] = tabla_frecuencias["Frec Rel"] * 100
tabla_frecuencias["Frec Ac"] = tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias["Frec Rel Ac"] = tabla_frecuencias["Frec Rel"].cumsum()
tabla_frecuencias["Frec Porc Ac"] = tabla_frecuencias["Frec Porcent"].cumsum()
print(tabla_frecuencias)
```
