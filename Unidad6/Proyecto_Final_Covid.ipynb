## 3. Gráficas

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Gráfico de barras horizontal por estado

```python
estados = df['ENTIDAD_NAC'].value_counts().head(20)
plt.figure(figsize=(10,6))
sns.barplot(y=estados.index, x=estados.values)
plt.title('Casos por estado (Top 20)')
plt.show()
```

### Histograma por edad

```python
plt.hist(df['EDAD'].dropna(), bins=30)
plt.title('Histograma de edad')
plt.show()
```

### Gráfico de líneas por fecha

```python
df['FECHA_INGRESO'] = pd.to_datetime(df['FECHA_INGRESO'], errors='coerce')
serie = df.groupby(df['FECHA_INGRESO'].dt.date).size()
plt.plot(serie)
plt.title('Evolución de casos')
plt.xticks(rotation=45)
plt.show()
```

### Barras apiladas por sexo y resultado

```python
tabla = pd.crosstab(df['SEXO'], df['RESULTADO_ANTIGENO'])
tabla.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Resultado por sexo')
plt.show()
```

### Circular por tipo de paciente

```python
df['TIPO_PACIENTE'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Tipo de paciente')
plt.ylabel('')
plt.show()
```

### Barras por comorbilidad

```python
comorb = ['DIABETES','HIPERTENSION','OBESIDAD','EPOC','ASMA','RENAL_CRONICA','CARDIOVASCULAR','TABAQUISMO']
conteo = df[comorb].sum().sort_values(ascending=False)
sns.barplot(x=conteo.values, y=conteo.index)
plt.title('Comorbilidades')
plt.show()
```

## 4. Probabilidades por comorbilidad

```python
total = len(df)
prob_simple = df[comorb].sum() / total
prob_simple
```

### Probabilidad condicional por estado

```python
cond = df.groupby('ENTIDAD_NAC')[comorb].sum()
p_cond = cond.div(df['ENTIDAD_NAC'].value_counts(), axis=0)
p_cond.head()
```

## 5. Distribución normal

```python
from scipy.stats import norm
p = norm.cdf(3000, 2360, 714) - norm.cdf(2000, 2360, 714)
p
```
